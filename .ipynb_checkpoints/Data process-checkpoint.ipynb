```python
import pandas as pd
import numpy as np
```

```python
raw = pd.read_csv("modbus audax.txt",header=None)
raw.head(30)
```

```python
raw.columns = ['Messy']
```

```python
raw.head()
```

```python
df = pd.DataFrame()
```

```python
#check separator
sep_col = raw['Messy'].str.startswith('----')  #.str
```

```python
#create groups
raw['groups'] = sep_col.cumsum() #this is the key!!!
```

```python
#filter separator rows
df = raw[~sep_col].copy()   #.copy()
# count ranking in each group
df['ranks'] = df.groupby('groups').cumcount()
df.head(30)
```

```python
df_new = df.pivot('groups','ranks','Messy')
```

```python
df_new
```

```python
df_new.to_csv("Cleaned_records.csv")
```
